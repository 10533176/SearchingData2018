```python
import re
import os
import numpy as np
from subprocess import run, PIPE
import plotly
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.graph_objs import *
import plotly.offline as offline

# plotly.tools.set_credentials_file(username='ajc10162046', api_key='7Y0EYySpaW1YeNOPAird')

names = ["jelinek_mercer", "dirichlet_prior", "absolute_discount", "LSI", "tfidf", "bm25"]

files = sorted([f for f in os.listdir('run_retrieval')if re.match('|'.join(names), f)])

if files[0] == '.DS_Store':
    files.remove('.DS_Store')

methods = ["ndcg_cut_10", "map_cut_1000", "P_5", "recall_1000"]

for method in methods:
    dic = {}

    for file in files:
        result = run('./../trec_eval/trec_eval -m all_trec ap_88_89/qrel_test run_retrieval/' + file + ' -q | grep -E "^' + method + '\s"', shell=True, stdout=PIPE, stderr=PIPE, universal_newlines=True)
        scores = [float(line.split('\t')[2]) for line in result.stdout.split('\n') if len(line.split('\t')) == 3]


        x = scores
        data = [go.Histogram(x=x)]
        layout = go.Layout(
            title = 'Scores of ' + file + ', ' + method,
            xaxis = dict(title= method + ' values'),
            yaxis = dict(title= 'frequency'))
        fig = go.Figure(data=data, layout=layout)
#         py.iplot(fig, filename= file + ', ' + method)
        offline.iplot(fig, filename = file + ', ' + method, image = 'png')

        score = sum(scores)/len(scores)
        param = re.search(re.compile('\d.\d*'), file).group()
        lm = re.search(re.compile('[a-z_a-z]*'), file).group()
        # print(lm, score, param)
        
        if lm not in scores:
            dic[lm] = {'best_param': param, 'score': score}
            continue
        if score > dic[lm]['score']:
            dic[lm]['best_param'] = param
            dic[lm]['score'] = score

    print(method)
    for lm in dic:
        print( lm, dic[lm])
    print('\n')

```
