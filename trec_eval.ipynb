```python
import os
import re
from subprocess import run, PIPE
import matplotlib.pyplot as plt
import numpy as np

files = sorted([f for f in os.listdir('run_retrieval')])
# print(files)
# when running it creates this .DS_Store file, must be removed for function to work properly
if files[0] == '.DS_Store':
    files.remove('.DS_Store')
    

methods = ["ndcg_cut_10", "map_cut_1000", "P_5", "recall_1000"]

for method in methods:
    results = []
    c = 0
    output = open('trec_result/' + method + '.txt', 'w')
    print(method)
    for file in files:
#         print(file)
        result = run('./../trec_eval/trec_eval -m all_trec ap_88_89/qrel_test run_retrieval/' + file + ' -q | grep -E "^' + method + '\s"', shell=True, stdout=PIPE, stderr=PIPE, universal_newlines=True)
#         print(result)
        scores = [float(line.split('\t')[2]) for line in result.stdout.split('\n')[:-2]] 
        score = sum(scores)/len(scores)
        if c == 0:
            queries = [line.split('\t')[1] for line in result.stdout.split('\n')[:-2]]
            output.write(method + '\t' + '\t'.join(queries) + '\n')
            c += 1

        print(file, score)
        output.write(file + '\t' + '\t'.join(str(x) for x in scores) + '\n')
    print("\n")
    output.close()
```
